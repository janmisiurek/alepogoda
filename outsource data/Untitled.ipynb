```python
import pandas as pd
import os
```

```python
dir_ = '/home/misiur/Projects/alepogoda/outsource data'
dir_list = os.listdir(dir_)

files = []
for element in dir_list:
    if element.endswith('.csv'):
        files.append(element)

dfs = []
for file in files:
    
    path = os.path.join(dir_, file)
    df = pd.read_csv(path)
    df.drop('ng-star-inserted 2', axis=1, inplace=True)
    df.rename(columns={'ng-star-inserted': 'temp_max', 'ng-star-inserted 3': 'temp_min'}, inplace=True)
    df = df.iloc[1:]
    
    month, year, city = file.split('-')
    year = int(year)
    month = int(month)
    city = city.strip('.csv').capitalize()
    days = df.index.tolist()
    date_column = pd.to_datetime([f'{day}-{month}-{year+2000}' for day in days], format='%d-%m-%Y')
    df.insert(0, 'Date', date_column)
    df.insert(0, 'City', city)
    dfs.append(df)
```

```python
for i, df in enumerate(dfs):
    print(files[i])
    print(df)
    print('------------------------------')
```

```python
for i, df in enumerate(dfs):
    path = f'/home/misiur/Projects/alepogoda/outsource data/historical/{files[i]}'
    df.to_csv(path, index=False)
```

```python
hist_temp_data = pd.concat(dfs)
print(hist_temp_data)
```
